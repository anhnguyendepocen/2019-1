* Error formulas for Riemann sums

```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's approximate the [Fresnel integral](https://en.wikipedia.org/wiki/Fresnel_integral)

$$
\int_0^1 \sin(x^2) dx
$$

using a Riemann sum with left endpoints. The error formula gives a bound on the error

$$
\left| \, \int_a^b f(x) dx - L_N(f) \, \right| \leq \frac{(b-a)^2 K_1}{2N}
$$

where $|f'(x)| \leq K_1$ for all $x \in [a,b]$. Let's plot $f'(x) = 2x\cos(x^2)$ to see a bound $|f'(x)| < K_1$.

```python
x = np.linspace(0,1,100)
y = 2*x*np.cos(x**2)
plt.plot(x,y)
plt.grid(True)
plt.show()
```

We can plot closer to the maximum near $x=0.8$.

```python
x = np.linspace(0.7,0.9,100)
y = 2*x*np.cos(x**2)
plt.plot(x,y)
plt.grid(True)
plt.show()
```

Since $|f'(x)| < 1.29$, we can find a $N$ such that

$$
\frac{(b-a)^2 K_1}{2N} < 0.001
$$

$$
\frac{1.29}{2(0.001)} < N
$$

```python
1.29/2/0.001
```

Let's approximate the integral.

```python
a = 0
b = 1
N = 645
x = np.linspace(a,b,N+1)
x_left = x[:N]
dx = (b - a)/N
y = np.sin(x_left**2)*dx
I = np.sum(y)
print(I)
```

Let's approximate the same integral but now with midpoints. The error formula gives a bound on the error

$$
\left| \, \int_a^b f(x) dx - M_N(f) \, \right| \leq \frac{(b-a)^3 K_2}{24N^2}
$$

where $|f''(x)| \leq K_2$ for all $x \in [a,b]$. Let's plot $f''(x) = 2\cos(x^2) - 4x^2\sin(x^2)$ to see a bound $|f''(x)| < K_2$.

```python
x = np.linspace(0,1,100)
y = 2*np.cos(x**2) - 4*x**2*np.sin(x**2)
plt.plot(x,y)
plt.grid(True)
plt.show()
```

The value at $x=1$ gives the maximum value.

```python
np.abs(2*np.cos(1) - 4*np.sin(1))
```

We get a much smaller $N$ value.

```python
np.sqrt(2.2852793274953065/24/0.001)
```

```python
a = 0
b = 1
N = 10
x = np.linspace(a,b,N+1)
x_midpoint = (x[1:] + x[:N])/2
dx = (b - a)/N
y = np.sin(x_midpoint**2)*dx
I = np.sum(y)
print(I)
```

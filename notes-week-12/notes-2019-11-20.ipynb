* Euler's method for second order equations

```python
import numpy as np
import matplotlib.pyplot as plt
```

Consider a second order differential equation with constant coefficients

$$
ay'' + by' + cy = f(t) \ , \ y(0)=y_0 \ , \ y'(0)=v_0
$$

How do we implement Euler's method to approximate solutions? Apply Euler's method to both $y$ and $y'$ simultaneously:

\begin{align*}
y_{n+1} & = y_n + y'_n h \\
y_{n+1}' & = y_n' + y''_n h
\end{align*}

where $y''_n = (f(t_n) - cy_n - by'_n)/a$.

Consider the equation $y'' + y = 0$, $y(0)=1$, $y'(0)=0$. We know the solution is $y(t) = \cos t$ and so we can compare our approximation with the exact solution.

```python
y0 = 1
v0 = 0
tf = 3*np.pi
steps = 100
N = steps + 1
t = np.linspace(0,tf,N)
h = tf/steps
y = np.zeros(N)
y[0] = y0
dy = np.zeros(N)
dy[0] = v0
for n in range(0,N-1):
    y[n+1] = y[n] + dy[n]*h
    dy[n+1] = dy[n] + (-y[n])*h
plt.plot(t,y,'b.-')
plt.plot(t,np.cos(t),'r')
plt.legend(["Euler's method","Exact solution"])
plt.show()
```

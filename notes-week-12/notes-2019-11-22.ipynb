* scipy.odeint

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi
```

## First order equations

Approximate the solution of $y' = -y$, $y(0) = 1$. We know $y(t) = e^{-t}$.

```python
def f(y,t):
    return -y

t = np.linspace(0,2,100)
y0 = 1
y = spi.odeint(f,y0,t)
```

```python
type(y)
```

```python
y.shape
```

```python
plt.plot(t,y,'b.')
plt.plot(t,np.exp(-t),'r')
```

Approximate the solution of $y' = \sin(y^2) + \cos(t)$ for various initial values.

```python
def f(y,t):
    return np.sin(y**2) + np.cos(t)

t = np.linspace(0,10,500)
for y0 in np.arange(-2,2,0.2):
    y = spi.odeint(f,y0,t)
    plt.plot(t,y,'b')
plt.show()
```

## First order systems

```python
def f(u,t):
    dudt = np.zeros(4)
    dudt[0] = u[1]
    dudt[1] = u[0]**2 - u[2]
    dudt[2] = u[3]
    dudt[3] = u[2] - u[0]
    return dudt

t = np.linspace(0,3,100)
u0 = [1,-1,2,0]
u = spi.odeint(f,u0,t)
```

Plot $x$ versus $y$.

```python
plt.plot(u[:,0],u[:,2])
```

```python
def f(u,t):
    dudt = np.array([0.0,0.0])
    dudt[0] = u[0]**2 - u[1]
    dudt[1] = u[1] + u[0] - 1
    return dudt

t = np.linspace(0,1,1000)
for _ in range(0,100):
    u0 = np.random.rand(2)
    u = spi.odeint(f,u0,t)
    plt.plot(u[:,0],u[:,1],'b')
plt.show()
```

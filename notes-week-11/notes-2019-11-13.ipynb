* Slope fields (aka direction fields)

```python
import numpy as np
import matplotlib.pyplot as plt
```

Most differential equations are impossible to solve exactly. Instead, we can always use numerical and graphical methods to study solutions of equations. The [slope field](https://www.jirka.org/diffyqs/html/slopefields_section.html) of a first order differential equation $y'=f(t,y)$ is a graphical method for visualizing solutions. The equation $y'=f(t,y)$ gives us complete information about the slopes of solutions at any point even if we don't know a formula for the solution. The procedure to create a slope field is simple: draw a small line of slope $f(t,y)$ at various points $(t,y)$ in a grid in the ty-plane.

```python
def f(t,y):
    return y - t

h = 0.1
L = 0.025
t_grid = np.arange(0,2+h,h)
y_grid = np.arange(-1,1+h,h)
for t in t_grid:
    for y in y_grid:
        m = f(t,y)
        theta = np.arctan(m)
        plt.plot([t - L*np.cos(theta),t + L*np.cos(theta)],
                 [y - L*np.sin(theta),y + L*np.sin(theta)],'b')
plt.grid(True)
plt.axis([0,2,-1,1])
plt.show()
```
